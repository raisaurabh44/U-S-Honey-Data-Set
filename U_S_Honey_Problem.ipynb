Q2. Which are the top 5 Honey producing state?



```python
new_df=df.groupby('state').sum()['production'].sort_values(ascending=False).reset_index()
new_df
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=new_df, x='state', y='production')
plt.xticks(rotation=90)
plt.show()
```

1	NorthDakota	513742000

2	California	423876000

3	SouthDakota	355726000

4	Florida	280934000

5	Montana	197173000

Top 5 staes with highest honey productions


Q3.What is the Change in mean Average price of Honey from 1995 to 2021.



```python
new_df= df.groupby('year').mean()['average_price']
new_df
```

```python
plt.plot(new_df)
```

From 1995 to 2017 the price of honey was increasing but we can see there was a sudden decrease in the price from 2018 and this trend keep going on constant.

Q4. Which was the year when production of Honey in whole US was the highest?


```python
new_df=df.groupby('year').sum()['production'].sort_values(ascending=False).reset_index()
new_df
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data=new_df, x='year', y='production')
plt.xticks(rotation=90)
plt.show()
```

The production of Honey in US was highest in year 2000


Q5.from the above inference we get the production was highest in the year 2000, now let infer which state was having highest contribution in that year?


```python
new_df=df[df['year']==2000].sort_values(by='production',ascending=False)
new_df
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data=new_df,x='state',y='production')
plt.xticks(rotation=70)
plt.show()
```
